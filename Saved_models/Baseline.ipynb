```python
# Now converting the label object using dummies
df_one_hot = pd.get_dummies(df,dtype=int) # Else it will give bool by defualt
df_one_hot.head()
```

```python
X = df_one_hot.drop('Rented_Bike_Count',axis = 1)
y = df_one_hot['Rented_Bike_Count']
```
